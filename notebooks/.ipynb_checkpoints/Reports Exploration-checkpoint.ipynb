```python
reports = pd.read_csv(path + 'surgeapp_reports.raw.csv')

import warnings
with warnings.catch_warnings():
    warnings.filterwarnings('ignore', r'call to matplotlib.use() has no effect')
import pandas_profiling as pp
pp.ProfileReport(reports)
```

```python
# Let's check out the reporting users, anyone who reports too frequently should maybe be ignored?
reporting_users = reports[['reporting_user_id', 'id']].groupby('reporting_user_id') \
    .count().reset_index().rename(columns={'id': 'c'})
print(reporting_users[reporting_users.c > 1].sort_values('c', ascending=False))

# Yeah, user '2767460' is crazy, so let's arbitrarily define anyone who reports more than 4 standard deviations above
# the mean as a crazy person and ignore their reports
crazy_users = reporting_users[reporting_users.c > reporting_users.c.mean() + reporting_users.c.std()*4]
print(crazy_users.sort_values('c', ascending=False))

reports = reports[reports.reporting_user_id.map(lambda x: x not in crazy_users.reporting_user_id)]
```
