# Account table
In this notebook we're going to read in the raw data exported from the accounts table and see if we can create any new variables from it

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
print(pd.__version__)
print(matplotlib.__version__)

path = '../data/'
```

### Read and summarize
Using the Pandas Profiling library, let's take a quick look at what's there in the raw data


```python
accounts = pd.read_csv(path + 'surgeapp_accounts.raw.csv')

import warnings
with warnings.catch_warnings():
    warnings.filterwarnings('ignore', r'call to matplotlib.use() has no effect')
import pandas_profiling as pp
pp.ProfileReport(accounts)
```

This gives a great overview at a glance of a lot of characteristics we're interested in.

We'll start our feature engineering work with the email address: let's extract the email provider and then use the most common values as new parameters

```python
accounts['email_provider'] = accounts.email.map(lambda x: x[x.index('@')+1:x.find('.', x.index('@')+1)])
accounts.head()
```

```python
pp.ProfileReport(accounts[['email', 'email_provider']])
```

Find the top 3 email providers

```python
topproviders = pd.value_counts(accounts.email_provider, sort=True).head(3).index
topproviders
```

**Change everything that isn't one of the top providers to "other"**

```python
accounts.loc[~accounts.email_provider.isin(topproviders), 'email_provider'] = 'other'
accounts[accounts.email_provider == 'other'].head()
```

```python
accounts['has_fb_id'] = pd.notnull(accounts.fb_id)
accounts.head()
```

```python
providers = pd.value_counts(accounts.email_provider, sort=True)
providers
```

```python
providers.plot.barh()
plt.show()
```

### Now let's look at the device

```python
accounts.loc[pd.isnull(accounts.device), 'device'] = 'unknown'
accounts.loc[accounts.device == '?unrecognized?', 'device'] = 'unknown'
accounts['device'] = accounts.device.str.lower()
accounts['device'] = accounts.device.str.replace(' ', '').str.replace('plus', '').str.replace('s', '').str.replace('c', '').str.replace('touh', 'touch')
devices = pd.value_counts(accounts.device, sort=True)
devices.plot.barh()
plt.show()
```

**Lots of values with very little representation, let's combine by device type...**

```python
accounts.loc[accounts.device.str.contains('iphone'), 'device'] = 'iphone'
accounts.loc[accounts.device.str.contains('ipad'), 'device'] = 'ipad'
accounts.loc[accounts.device.str.contains('ipod'), 'device'] = 'ipod'
devices = pd.value_counts(accounts.device, sort=True)
devices.plot.barh()
plt.show()
```

**Let's take one last look at our dataframe before we finish it up and write it to file...**

```python
cols = ['id', 'device', 'email_provider', 'has_fb_id']
accounts = accounts[cols]
pp.ProfileReport(accounts)
```

```python
accounts.to_csv(path + 'surgeapp_accounts.csv', index=False)
```
